```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train_data = pd.read_csv('/home/asavari/Downloads/Gotem Pumpkins.csv')
test_data = pd.read_csv('/home/asavari/Downloads/Freyja_Pumpkins.csv')
```

```python
# The last column is the target variable and the rest are features
X_train = train_data.iloc[:, :-1].values
y_train = train_data.iloc[:, -1].values
X_test = test_data.iloc[:, :-1].values
y_test = test_data.iloc[:, -1].values
```

```python
# Normalization
X_train_min = X_train.min(axis=0)
X_train_max = X_train.max(axis=0)
X_train = (X_train - X_train_min) / (X_train_max - X_train_min)
X_test = (X_test - X_train_min) / (X_train_max - X_train_min)
```

```python
unique_classes = np.unique(y_train)
class_to_num = {cls: i for i, cls in enumerate(unique_classes)}
num_to_class = {i: cls for cls, i in class_to_num.items()}

y_train_num = np.array([class_to_num[cls] for cls in y_train])
y_test_num = np.array([class_to_num[cls] for cls in y_test])
```

```python
# Sigmoid function
def sigmoid(z):
    z = np.clip(z, -500, 500)  # Clipping to prevent overflow
    return 1 / (1 + np.exp(-z))
```

```python
# Define cost function
def cost_function(X, y, weights):
    m = X.shape[0]
    predictions = sigmoid(np.dot(X, weights)) 
    epsilon = 1e-15  # Small value to prevent log(0)
    predictions = np.clip(predictions, epsilon, 1 - epsilon)
    cost = -(1/m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost
```

```python
# Gradient descent
def gradient_descent(X, y, weights, learning_rate, iterations):
    m = X.shape[0]
    costs = []
    for i in range(iterations):
        gradient = (1/m) * np.dot(X.T, (sigmoid(np.dot(X, weights)) - y)) 
        weights -= learning_rate * gradient
        cost = cost_function(X, y, weights)
        costs.append(cost)
    return weights, costs
```

```python
X_train_bias = np.hstack((np.ones((X_train.shape[0], 1)), X_train))

# Initialize weights
weights = np.zeros((X_train_bias.shape[1], len(unique_classes)))

# Training the model
learning_rate = 0.1
iterations = 1000
all_costs = []

for i in range(len(unique_classes)):
    y_binary = (y_train_num == i).astype(int)
    weight, costs = gradient_descent(X_train_bias, y_binary, np.zeros(X_train_bias.shape[1]), learning_rate, iterations)
    weights[:, i] = weight
    all_costs.append(costs)
```

```python
for i, costs in enumerate(all_costs):
    plt.figure(figsize=(8, 5))
    plt.plot(range(iterations), costs, label=f"Class: {num_to_class[i]}", color="blue" if i == 0 else "orange")
    plt.xlabel("Iterations")
    plt.ylabel("Cost")
    plt.title(f"Cost Convergence for Class: {num_to_class[i]}")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Add bias term to X_test
X_test_bias = np.hstack((np.ones((X_test.shape[0], 1)), X_test))

# Predict
probabilities = sigmoid(X_test_bias @ weights)
predicted_classes = np.argmax(probabilities, axis=1)
predicted_labels = [num_to_class[i] for i in predicted_classes]

# Comparing with actuallabels
actual_labels = [num_to_class[i] for i in y_test_num]
```

```python
#Accuracy
accuracy = np.mean(predicted_classes == y_test_num)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
# Unique classes
unique_classes = list(set(actual_labels))

# Initialize confusion matrix
conf_matrix = np.zeros((len(unique_classes), len(unique_classes)), dtype=int)

# Populate confusion matrix
for true_label, pred_label in zip(actual_labels, predicted_labels):
    true_idx = unique_classes.index(true_label)
    pred_idx = unique_classes.index(pred_label)
    conf_matrix[true_idx, pred_idx] += 1

print("Confusion Matrix:")
print(conf_matrix)
```

```python
def plot_confusion_matrix(conf_matrix, classes):
    plt.figure(figsize=(8, 6))
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title("Confusion Matrix", fontsize=16)
    plt.colorbar()

    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, fontsize=12)
    plt.yticks(tick_marks, classes, fontsize=12)

    # Adding values inside the matrix
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            plt.text(j, i, conf_matrix[i, j], horizontalalignment="center", color="white" if conf_matrix[i, j] > conf_matrix.max()/2 else "black")

    plt.ylabel("Actual Label", fontsize=14)
    plt.xlabel("Predicted Label", fontsize=14)
    plt.tight_layout()
    plt.show()


plot_confusion_matrix(conf_matrix, unique_classes)
```
