```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
train_data = pd.read_csv('/home/asavari/Downloads/Train.csv')
test_data = pd.read_csv('/home/asavari/Downloads/Test.csv')
```

```python
train_data = train_data.drop(["Unnamed: 0",'Name'],axis=1)
train_data = pd.get_dummies(train_data, columns=['Location','Fuel_Type','Transmission','Owner_Type'],drop_first ='True')
train_data = train_data.dropna()
train_data.Mileage = train_data.Mileage.str.replace(' kmpl' , ' ').astype(float)
train_data.Engine = train_data.Engine.str.replace(' CC', ' ').astype(int)
train_data['Power'] = train_data['Power'].replace('null bhp', '0 bhp')
train_data.Power = train_data.Power.str.replace(' bhp',' ').astype(float)
train_data
```

```python
test_data = test_data.drop(['Unnamed: 0', 'Name'], axis=1)
test_data = pd.get_dummies(test_data, columns=['Location', 'Fuel_Type', 'Transmission', 'Owner_Type'], drop_first=True)
test_data = test_data.dropna()
test_data.Mileage = test_data.Mileage.str.replace(' kmpl', '').astype(float)
test_data.Engine = test_data.Engine.str.replace(' CC', '').astype(float)
test_data['Power'] = test_data['Power'].replace('null bhp', '0 bhp')
test_data.Power = test_data.Power.str.replace(' bhp', ' ').astype(float)
test_data
```

```python
features = ['Year', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats']
X_train = train_data[features].values
y_train = train_data['Price'].values
X_test = test_data[features].values
y_test_actual = test_data['Price'].values  #for accuracy stuff
```

```python
# Normalization code
X_train_min = X_train.min(axis=0)
X_train_max = X_train.max(axis=0)
X_train = (X_train - X_train_min) / (X_train_max - X_train_min)
X_test = (X_test - X_train_min) / (X_train_max - X_train_min)
```

```python
print("Shape of X_standardized:", X_train.shape)
print("Shape of Y_standardized:", y_train.shape)
```

```python
# gradient descent
def gradient_descent(x, y, w, b, a, iterations):
    x = np.array(x, dtype=np.float64)
    y = np.array(y, dtype=np.float64)
    w = np.array(w, dtype=np.float64)

    m = len(y) 
    cost_history = []

    for i in range(iterations):
        predictions = x.dot(w) + b
        errors = predictions - y
        w -= a * (x.T.dot(errors) / m)
        b -= a * (errors.sum() / m)

        # Calculate and store the cost
        cost = ((errors ** 2).sum()) / (2 * m)
        cost_history.append(cost)

        # Output metrics every 100 iterations
        if i % 100 == 0:
            mse = (errors ** 2).mean()
            rmse = mse ** 0.5
            r2 = 1 - (errors.var() / y.var())
            print(f"Iteration {i}: MSE={mse:.2f}, RMSE={rmse:.2f}, R²={r2:.4f}, Cost={cost:.2f}")

    return w, b, cost_history
```

```python
#cost
def cost(x,y,w,a,b):
    m,n= x.shape
    predictions = x.dot(w) + b  
    cost_value = (1 / (2 * m)) * np.sum((predictions - y) ** 2) + (a / (2 * m)) * np.sum(w ** 2)  
    return cost_value
```

```python
# acc
def acc (y_true,y_pred):
    mse =((y_true- y_pred)**2).mean()
    rmse = mse**0.5
    mean_y = y_true.mean()
    total =((y_true - mean_y)**2).sum()
    res = ((y_true -y_pred) **2).sum()
    r2 = 1 - (res/total)
    
    return mse,rmse,r2
```

```python
#lr
def reg(x,y,lr,epochs):
    x = np.array(x, dtype=np.float64)  
    y = np.array(y, dtype=np.float64)  
    n = x.shape[1] 
    w = np.zeros(n, dtype=np.float64) 
    b = 0.0  

    w,b,cost_history = gradient_descent(x,y,w,b,lr,epochs)
    return w,b,cost_history


lr = 0.001
epochs = 1000
w,b, cost_history = reg(X_train,y_train,lr, epochs)

print("\nFinal cost:", cost_history[-1])
print("Final weights:", w)
print("Final bias:", b)
```

```python
#mean square scores

train_predictions = X_train.dot(w) + b
test_predictions = X_test.dot(w) + b

_, _, r2_train = acc(y_train, train_predictions)
_, _,r2_test = acc(y_test_actual, test_predictions)

# Print R² scores
print(f"Training R² Score: {r2_train * 100:.2f}%")
print(f"Testing R² Score: {r2_test * 100:.2f}%")
```

```python
#graph1; cost over epochs

cost_df = pd.DataFrame({
    'Iteration': np.arange(len(cost_history)),  # Create an array of iteration numbers
    'Cost': cost_history  # Cost values
})

sns.set(style="whitegrid")
# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=cost_df, x='Iteration', y='Cost', linewidth=2.5, markersize=5)


plt.title('Cost over Iterations', fontsize=16)
plt.xlabel('Iteration', fontsize=14)
plt.ylabel('Cost', fontsize=14)
plt.legend()

plt.show()
```
